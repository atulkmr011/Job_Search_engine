```python
# Importing essential libraries for performing Natural Language Processing on 'fake_train' dataset
import nltk
import re
nltk.download('stopwords')
from nltk.corpus import stopwords #for removing the stopwords
from nltk.stem import PorterStemmer #for stemming
from nltk.stem import WordNetLemmatizer #For lemmetizer
```

```python
len(result) 
  
finaldata = result[["code","_id","title","title_index","filename_key","filename_value"]]          # Required columns - Title and combined data  
finaldata = finaldata.set_index('title_index')    # Setting the job title title as index 
  
finaldata.head(10) 
finaldata["filename_value"][0] 
```

```python
pattern='[^a-zA-Z]'
finaldata["filename_value"] = finaldata["filename_value"].apply(lambda x: re.sub(pattern," ",str(x)))
```

```python
#takes time
lm = WordNetLemmatizer()
def preprocess_sentences(text):
    text = text.lower()
    temp_sent =[]
    words = text.split()
    words = nltk.word_tokenize(text)
    words = [word for word in words if word not in set(stopwords.words('english'))]
    temp_sent = [lm.lemmatize(word) for word in words]
    finalsent = ' '.join(temp_sent)
    return finalsent
        
    
finaldata["filename_value_processed"]= finaldata["filename_value"].apply(preprocess_sentences) 
finaldata.head()    
    
```

```python
#Indexing for the title
indices_sig = pd.Series(dfdf.index, index=dfdf['filename_key']).drop_duplicates()
```

```python
indices_sig
```

```python
#function for recommendation
def give_rec(title, sig=sig):
    # Get the index corresponding to original_title
    idx = indices_sig[title]

    # Get the pairwsie similarity scores 
    sig_scores = list(enumerate(sig[idx]))

    # Sort the output 
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Scores of the  most similar output
    sig_scores = sig_scores[1:2064]

    #  indices
    title_indices = [i[0] for i in sig_scores]

    # Top 10 most similar 
    return dfdf['filename_key'].iloc[title_indices]
```
